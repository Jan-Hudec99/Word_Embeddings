```python
import tensorflow_hub as hub
import tensorflow.compat.v1 as tf
import pandas as pd
import os
import pickle
import time

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
tf.disable_eager_execution()

elmo = hub.Module("https://tfhub.dev/google/elmo/3", trainable = True)

dataset = pd.read_csv("IMDB Dataset.csv")
```

```python
# getting the vectors for 10000 reviews in batches
vectors = []
i = 10
start_time = time.time()
while i <= 10000:
    embeddings = elmo(dataset['review'].iloc[i-10:i], signature = "default", as_dict = True)["elmo"]
    init = tf.initialize_all_variables()
    sess = tf.Session()
    sess.run(init)

    for j in sess.run(embeddings):
        vectors.append(j)

    i += 10
```

```python
# saving the vectors to disk
pickle_out = open("elmo3_vectors.pickle", "wb")
pickle.dump(vectors, pickle_out)
pickle_out.close()
```
